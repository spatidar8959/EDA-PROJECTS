## 3. ***Data Wrangling***

Data Wrangling Code

---



First : we drop all the unneccessary columns from the dataframe.
Second : we drop all the null values from the dataframe.
Third : we convert the type of reservation_status_date column.
Fourth : we print all the unique values from the unique columns.

````python
# Write your code to make your dataset analysis ready.
#First : we drop all the unneccessary columns from the dataframe.
hotel_df.drop(['company','agent'],axis = 1,inplace=True)
````

````python
# Second : we drop all the null values from the dataframe.
hotel_df.dropna(inplace= True)
hotel_df.isna().sum()
````

````python
# Third : we convert the type of reservation_status_date column.
hotel_df['reservation_status_date'] = pd.to_datetime(hotel_df['reservation_status_date'])
hotel_df['reservation_status_date'].info()
````

````python
# fourth : we print all the unique values from the unique columns.
for col in hotel_df.describe(include='object').columns:
  print(col)
  print(hotel_df[col].unique())
  print('-'*50)
````

````python
# FIfth : we remove all the outliers in the adr column.
plt.figure(figsize = (6,3))
hotel_df['adr'].plot.box()
````

````python
# REmoving outliers
hotel_df = hotel_df[hotel_df['adr']<5000]
````

````python
# Chart - 11 visualization code
# Calculate cancelled dataframe with adr
cancelled_df_adr = cancelled_df.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

# Calculate not cancelled dataframe with adr
not_cancelled_data = hotel_df[hotel_df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date' , inplace = True)

# visualizing cancelled adr df and not cancelled adr df with line chart
plt.figure(figsize = (20,6))
plt.title('AVERAGE DAILY RATE')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'NOT CANCELLED')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'CANCELLED')
plt.legend()
````
